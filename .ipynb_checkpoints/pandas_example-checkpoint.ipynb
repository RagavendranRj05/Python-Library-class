```python
sample = pd.read_csv("SampleSuperstore.csv")
sample
```

```python
samp = sample.head(21)
samp
```

```python
samp.groupby("City")["Sales"].mean()
```

```python
sample.groupby("City")["Sales"].mean()
```

```python
samp.groupby("City")["Profit"].mean()  # aggregate function (sum,mean,min,max,count,etc)
```

```python
samp.groupby("City")["Profit"].agg(["mean","sum","min","max"])
```

```python
samp.groupby("Country")["City"].count()  # aggregate function (sum,mean,min,max,count,etc)
```

```python
samp.groupby("City")["Country"].count()
```

```python
samp.groupby("State")["City"].count()
```

```python
samp.groupby("Region")["City"].count()
```

```python
samp.groupby(["State","City"])["Country"].count()
```
