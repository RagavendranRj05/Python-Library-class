## MatplotLib 

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 13, 100)

plt.subplot(3, 3, 1)  # first plot (top-left)
plt.plot(x, np.sin(x))
plt.title('Plot 1')

plt.subplot(3, 3, 2)  # second plot (top-right)
plt.plot(x, np.cos(x))
plt.title('Plot 2')

plt.subplot(2, 2, 3)  # first plot (top-left)
plt.plot(x, np.sin(x))
plt.title('Plot 1')

plt.subplot(2, 2, 4)  # second plot (top-right)
plt.plot(x, np.cos(x))
plt.title('Plot 2')

  # adjust spacing so titles don't overlap
plt.show()
```

```python
## Creating Subplots

plt.subplot(3,3,1)
plt.plot(x,y,'rX:')
plt.subplot(3,3,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'gs')
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y,'co') 
plt.show() 
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(3, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.xlabel('xvalues')
plt.ylabel('yvalues')
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [1,5,8] 
y = [11,16,9]  

x2 = [3,9,15] 
y2 = [6,15,7] 
plt.bar(x, y,color='m') 
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis') 
 
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 120) for std in range(1, 5)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False);  
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [10, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.0, 0.10, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('square')
plt.show()
```
