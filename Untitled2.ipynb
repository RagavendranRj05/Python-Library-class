```python
login = {
    "admin": "1234"
}
 
while True:
    username = input("Enter username: ")
    password = input("Enter password: ")
 
    if username in login and login[username] == password:
        print("Login successful")
        print("Welcome", username)
        break
    else:
        print("Wrong username or password")
        print("Try again\n")
 
```
