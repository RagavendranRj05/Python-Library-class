## MatplotLib 

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,13)
x
```

```python
y = np.arange(11,23)
y
```

```python
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple plot")
plt.savefig("Simple-plot.jpg")
```

```python
plt.scatter(x,y,c="k")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple plot")
```

```python
plt.scatter(x,y,c="m",marker="*",s=200,edgecolor="gold",alpha=0.7)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Simple plot")
```

```python
a = np.array([1,2,3,4,5,6])
a
```

```python
b = np.array([10,20,30,40,50,60])
b
```

```python
plt.plot(a,b,marker="p",color = "blue",linestyle=":",markersize=10)
plt.xlabel("number of chocolates",color="red")
plt.ylabel("Chocolate price")
plt.title("Dairy Milk Chocolate")
plt.show()
```

```python
colours=["red","orange","pink","violet","m","cyan"]
```

```python
plt.scatter(a,b,marker="*",s=200,c=colours)
```

```python
plt.figure(facecolor="yellow")
plt.plot(a,b,marker="p",color = "White",linestyle=":",markersize=10)
plt.gca().set_facecolor('pink')
plt.xlabel("number of chocolates")
plt.ylabel("Chocolate price")
plt.title("Dairy Milk Chocolate")
plt.show()
```

```python
plt.figure(facecolor="yellow")
plt.plot(a,b,marker="p",color = "White",linestyle=":",markersize=10)
plt.gca().set_facecolor('pink')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("number of chocolates",fontsize = 18 )
plt.ylabel("Chocolate price",fontsize=16)
plt.title("Dairy Milk Chocolate",fontsize=18)
plt.show()
```

```python
plt.figure(facecolor="yellow")
plt.plot(a,b,marker="p",color = "White",linestyle=":",markersize=10,lw = 0.9)
plt.gca().set_facecolor('pink')
plt.xticks(fontsize=16,color="red")
plt.yticks(fontsize=16,color="m")
plt.xlabel("number of chocolates",fontsize = 18,color="red")
plt.ylabel("Chocolate price",fontsize=16,color="green")
plt.title("Dairy Milk Chocolate",fontsize=18,color="cyan")
plt.show()
```
